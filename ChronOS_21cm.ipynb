# Imports and Utility Definitions

```python
#@title Import
import py21cmfast as p21c
import numpy as np
import tools21cm as t2c

import warnings
warnings.filterwarnings("ignore")
from pathlib import Path
import os
import sys

# matplotlib imports
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import cm, colors
import ipywidgets as widgets
from ipywidgets import interact

# plotly imports and definitions
import plotly.graph_objects as go
from plotly.colors import sample_colorscale, get_colorscale, make_colorscale
from plotly.subplots import make_subplots
import plotly.express as px
def mincut_colorscale(scalename, min, max, cutoff, N=15):
  return make_colorscale(sample_colorscale(scalename, np.linspace(cutoff,max,N)/max, low=min, high=max))

def GET_Z_CONE_RANGE(z_min,z_max,dz=None,BOX_PARAM=50,BOX_LEN=50): # filenames will be dependent on BOX_PARAM AND BOX_LEN
  z_full_range = np.round(t2c.lightcone.redshifts_at_equal_comoving_distance(5,35,BOX_PARAM,BOX_LEN),4)
  z_range = np.array([z for z in z_full_range if z_min <= z <= z_max])
  if dz is None:
    return z_range
  # indices = range(0,len(z_range),max(1,(z_range[-1]-z_range[0])//step))
  z_linspace = np.linspace(z_min,z_max,int(max(1,(z_max-z_min)//dz)))
  indices = []
  for zx in z_linspace:
    for i, z in enumerate(z_range):
      if zx-z < 0:
        indices.append(i)
        break
  return z_range[indices]

def get_brightness_temp_box(box, box_len, redshift, hubble, matter, baryon, save=False):
    filename = f"dTb_BOX_PARAM_{box}_Z_PARAM_{redshift}.npy"

    if os.path.exists(filename):
      print(f"Loading {filename}.")
      return np.load(filename)

    initial_conditions = p21c.initial_conditions(
        # HII_DIM is the number of cells along the length of a box
        # box_len is the total length of a side of the box
        user_params={"HII_DIM": box, "BOX_LEN": box_len, "N_THREADS": 2},
        cosmo_params=p21c.CosmoParams(
            SIGMA_8=0.8,
            hlittle=hubble,
            OMm=matter,
            OMb=baryon
        ),
        random_seed=54321
    )

    perturbed_field = p21c.perturb_field(
        redshift=redshift,
        init_boxes=initial_conditions
    )

    ionized_field = p21c.ionize_box(
        perturbed_field=perturbed_field
    )

    brightness_temp = p21c.brightness_temperature(
        ionized_box=ionized_field,
        perturbed_field=perturbed_field
    )

    dTb_box = brightness_temp.brightness_temp
    if save: np.save(filename, dTb_box)
    # print(brightness_temp.astro_params)
    # print(brightness_temp.user_params)
    # Return the raw numpy array
    return dTb_box
```

# Simulation

```python
#@title Box Simulation
BOX_PARAM = 50
BOX_LEN = BOX_PARAM
Z_STEP = 0.2
Z_PARAM_RANGE = GET_Z_CONE_RANGE(7,12+1,Z_STEP)
Z_validate = lambda z: z_val if (z_val:=np.ceil(z/Z_STEP)*Z_STEP) in Z_PARAM_RANGE else Z_PARAM_RANGE[-1]

# 1. Simulate and return a numpy array of a box of size NxNxN, N=BOX_PARAM
dTb_array = {
  Z_PARAM: get_brightness_temp_box(
    box=BOX_PARAM,
    box_len=BOX_LEN,
    redshift=Z_PARAM,
    hubble=0.69+(0 if print(f"Simulating z={Z_PARAM} ...") else 0),
    matter=0.31,
    baryon=0.02,
    save=True
) for Z_PARAM in Z_PARAM_RANGE}

dTb_max = np.nanmax([np.nanmax(dTb_array[z]) for z in Z_PARAM_RANGE])
dTb_min = np.nanmin([np.nanmin(dTb_array[z]) for z in Z_PARAM_RANGE])
```

```python
#@title Plot a specific slice of the box
def plot_tomographic_slice(box_data, box_len, redshift, slice_index, cmap="Turbo", slice_plane='z'):
    if not -box_data.shape[2] <= slice_index < box_data.shape[2]:
      raise ValueError(f"Slice index {slice_index} is out of bounds for axis of size {box_data.shape[2]}")

    # Select the specific slice
    if slice_plane.lower() == 'x':
      Tb_slice = box_data[slice_index, :, :]
    elif slice_plane.lower() == 'y':
      Tb_slice = box_data[:, slice_index, :]
    else:
      Tb_slice = box_data[:, :, slice_index]


    fig = go.Figure(data=go.Heatmap(
        z=Tb_slice,
        hoverinfo = "z",
        zmax = dTb_max,
        zmin = dTb_min,
        colorscale = cmap,
        ))

    fig.update_layout(
    yaxis_scaleanchor="x",
    width = 600, height=600,
    title=dict(
        text=f"Brightness Temperature Slice {slice_index} at Redshift z = {redshift}",
        y=0.91,
        x=0.5,
        xanchor="center",
        yanchor="middle"
    ),
    xaxis=dict(title=dict(text=r"x [Mpc]")),
    yaxis=dict(title=dict(text=r"y [Mpc]"))
    )
    fig.show()

    fig.update_layout(
    autosize=True,
    width=None,
    height=None
)
    fig.write_html(f"htmls/heatmap_slice_{slice_index}_z_{redshift}.html")

Z_PARAM = 10
Z_PARAM = Z_validate(Z_PARAM)
plot_tomographic_slice(
    box_data=dTb_array[Z_PARAM],
    box_len=BOX_PARAM,
    redshift=Z_PARAM,
    slice_index=0
)
```

```python
#@title Mean Brightness Temperature of the box, Slide through Redshift
def plot_tomographic_slice(boxes, box_len, redshift_range, cmap="Turbo", slice_plane='z'):
    # Select the specific slice
    def select_slice(box, index, plane):
      if plane.lower() == 'x':
        box_slice = box[index, :, :]
      elif plane.lower() == 'y':
        box_slice = box[:, index, :]
      else:
        box_slice = box[:, :, index]
      return box_slice

    def box_mean(box):
        mean = np.array([[np.mean(box[i, j, :]) for j in range(len(box[i]))] for i in range(len(box))])
        return mean

    maketitle = lambda index, z: f"Mean Brightness Temperature at redshift z={z}"

    fig = go.Figure()

    slice_index = 0
    for redshift in redshift_range:
      fig.add_trace(go.Heatmap(
        # z=select_slice(boxes[redshift], slice_index, slice_plane),
        z=box_mean(boxes[redshift]),
        hoverinfo = "z", # this is z-axis, not redshift
        zmax = dTb_max,
        zmin = dTb_min,
        colorscale = cmap,
    ))

    # fig.data[0].visible = True

    sliders = [dict(
    active=len(redshift_range)-1,
    # currentvalue={"prefix": "Frequency: "},
    pad={"t": 50},
    steps=[dict(
        method="update",
        args=[
            {"visible": [_z == z for _z in redshift_range]},
            {"title": dict(
                            text=maketitle(slice_index, z),
                            y=0.91,
                            x=0.5,
                            xanchor="center",
                            yanchor="middle"
                        )}
        ],
        label=f"z={z}",
    ) for z in redshift_range]
    )]

    fig.update_layout(
    yaxis_scaleanchor="x",
    width = 550, height=600,
    title=dict(
        text=maketitle(slice_index, redshift),
        y=0.91,
        x=0.5,
        xanchor="center",
        yanchor="middle"
    ),
    xaxis=dict(title=dict(text=r"x [Mpc]")),
    yaxis=dict(title=dict(text=r"y [Mpc]")),
    sliders=sliders
    )

    fig.show()
    return fig

fig = plot_tomographic_slice(
    boxes=dTb_array,
    box_len=BOX_PARAM,
    redshift_range=Z_PARAM_RANGE
)

fig.update_layout(
    autosize=True,
    width=None,
    height=None
)
fig.write_html("htmls/mean2d_brightness_temperature.html")
```

```python
#@title Plot a specific slice of the box
def plot_tomographic_slice(box_data, box_len, redshift, slice_index, cmap="Turbo", slice_plane='z'):
    if not -box_data.shape[2] <= slice_index < box_data.shape[2]:
      raise ValueError(f"Slice index {slice_index} is out of bounds for axis of size {box_data.shape[2]}")

    # Select the specific slice
    if slice_plane.lower() == 'x':
      Tb_slice = box_data[slice_index, :, :]
    elif slice_plane.lower() == 'y':
      Tb_slice = box_data[:, slice_index, :]
    else:
      Tb_slice = box_data[:, :, slice_index]

    fig, ax = plt.subplots(figsize=(7, 6))
    im = ax.imshow(
        Tb_slice,
        origin="lower",
        cmap=cmap,
        extent=[0, box_len, 0, box_len]
    )

    cbar = plt.colorbar(im, ax=ax)
    cbar.set_label(r"$\delta T_b$ [mK]")

    ax.set_xlabel("x [Mpc]")
    ax.set_ylabel("y [Mpc]")
    ax.set_title(rf"$z={redshift:.2f}$ | Slice: {slice_index} [max {box_len-1}]")

    plt.tight_layout()
    plt.show()

Z_PARAM = 10
plot_tomographic_slice(
    box_data=dTb_array[Z_PARAM],
    box_len=BOX_PARAM,
    redshift=Z_PARAM,
    slice_index=0
)
```

```python
#@title Interactive Tomography Slider
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
import numpy as np

def explore_tomography(box_data, box_len, redshift, slice_plane='z', cmap="Turbo"):
    """
    Creates an interactive slider to explore slices of a 3D box.
    """

    # 1. Calculate Global Color Limits
    # This ensures the color scale doesn't "jump" as you slide.
    # We calculate this ONCE for the whole box.
    global_vmin = np.nanmin(box_data)
    global_vmax = np.nanmax(box_data)

    # Determine max slices based on chosen plane
    if slice_plane.lower() == 'x':
        max_idx = box_data.shape[0] - 1
    elif slice_plane.lower() == 'y':
        max_idx = box_data.shape[1] - 1
    else: # z
        max_idx = box_data.shape[2] - 1

    # 2. Define the Update Function
    def update_plot(slice_idx):
        # Select the slice
        if slice_plane.lower() == 'x':
            Tb_slice = box_data[slice_idx, :, :]
            xlabel, ylabel = "y [Mpc]", "z [Mpc]"
        elif slice_plane.lower() == 'y':
            Tb_slice = box_data[:, slice_idx, :]
            xlabel, ylabel = "x [Mpc]", "z [Mpc]"
        else:
            Tb_slice = box_data[:, :, slice_idx]
            xlabel, ylabel = "x [Mpc]", "y [Mpc]"

        # Plotting
        fig, ax = plt.subplots(figsize=(7, 6))

        im = ax.imshow(
            Tb_slice,
            origin="lower",
            cmap=cmap,
            extent=[0, box_len, 0, box_len],
            vmin=global_vmin, # <--- CRITICAL: Use fixed limits
            vmax=global_vmax
        )

        cbar = plt.colorbar(im, ax=ax)
        cbar.set_label(r"$\delta T_b$ [mK]")

        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.set_title(rf"$z={redshift:.2f}$ | Slice: {slice_idx}/{max_idx} ({slice_plane}-axis)")

        plt.tight_layout()
        plt.show()

    # 3. Create the Slider
    # layout={'width': '500px'} makes the slider wider and easier to grab
    slider = widgets.IntSlider(
        min=0,
        max=max_idx,
        step=1,
        value=max_idx // 2,
        description='Slice:',
        layout={'width': '600px'}
    )

    # 4. Launch the Interaction
    interact(update_plot, slice_idx=slider)

# --- HOW TO RUN IT ---
Z_PARAM=10
explore_tomography(dTb_array[Z_PARAM], BOX_PARAM, Z_PARAM)
```

```python
#@title Interactive 4D Tomography (Redshift + Slice Sliders)
def visualize_4d_tomography(data_dict, box_len, slice_plane='z', cmap='Rainbow'):
    """
    Interactive 4D visualizer.
    data_dict: Dictionary {z: 3D_array}
    """

    # 1. SETUP GLOBAL NORM (Scan all redshifts for min/max)
    # This ensures colors don't flicker when changing Z or Slice
    all_values = np.concatenate([arr.flatten() for arr in data_dict.values()])
    vmin, vmax = np.nanpercentile(all_values, 1), np.nanpercentile(all_values, 99)
    norm = colors.Normalize(vmin=vmin, vmax=vmax)
    m = cm.ScalarMappable(cmap=cmap, norm=norm)
    m.set_array([])

    # Sorted redshifts for the slider
    z_keys = sorted(list(data_dict.keys()), reverse=True) # High z to Low z

    # 2. PLOT FUNCTION
    def plot_frame(z_val, slice_idx):
        box_data = data_dict[z_val]
        dim = box_data.shape[0]
        coord = np.linspace(0, box_len, dim)

        # --- A. YOUR PROJECTIONS (Index 0) ---
        # Note: Transposes (.T) aligned to match plotting orientation
        proj_z = box_data[:, :, 0].T   # Floor (xy plane)
        proj_y = box_data[:, 0, :].T   # Wall (xz plane)
        proj_x = box_data[0, :, :].T   # Wall (yz plane)

        fig = plt.figure(figsize=(10, 8))
        ax = fig.add_subplot(111, projection='3d')

        # --- B. DRAW THE WALL PROJECTIONS ---
        X_grid, Y_grid = np.meshgrid(coord, coord)

        # Floor (z=0)
        ax.contourf(X_grid, Y_grid, proj_z, zdir='z', offset=0,
                    cmap=cmap, levels=50, vmin=vmin, vmax=vmax, alpha=0.1)

        # Side Wall (x=0) - typically 'left/back'
        # We project proj_x onto the x=0 plane (yz axes)
        # Grid setup: Y (horizontal on wall), Z (vertical on wall)
        Y_wall, Z_wall = np.meshgrid(coord, coord)
        ax.contourf(proj_x, Y_wall, Z_wall, zdir='x', offset=0,
                    cmap=cmap, levels=50, vmin=vmin, vmax=vmax, alpha=0.1)

        # Back Wall (y=box_len) - typically 'right/back'
        # We project proj_y onto the y=Max plane (xz axes)
        X_wall, Z_wall = np.meshgrid(coord, coord)
        ax.contourf(X_wall, proj_y, Z_wall, zdir='y', offset=box_len,
                    cmap=cmap, levels=50, vmin=vmin, vmax=vmax, alpha=0.1)

        # --- C. DRAW THE MOVING SLICE (Solid Surface) ---
        if slice_plane == 'z':
            # Slice position
            z_pos = slice_idx * (box_len / dim)

            # Data for the slice
            slice_data = box_data[:, :, slice_idx].T

            # Create Surface Grid
            X_surf, Y_surf = np.meshgrid(coord, coord)
            Z_surf = np.full_like(X_surf, z_pos)

            # Plot Surface
            ax.plot_surface(X_surf, Y_surf, Z_surf,
                            facecolors=m.to_rgba(slice_data),
                            shade=False, rstride=1, cstride=1, antialiased=False)

            # Slice Outline
            ax.plot([0, box_len, box_len, 0, 0],
                    [0, 0, box_len, box_len, 0],
                    [z_pos, z_pos, z_pos, z_pos, z_pos], 'k', lw=1)
        elif slice_plane == 'y':
            y_pos = slice_idx * (box_len / dim)
            slice_data = box_data[:, slice_idx, :].T
            X_surf, Z_surf = np.meshgrid(coord, coord)
            Y_surf = np.full_like(X_surf, y_pos)

            ax.plot_surface(X_surf, Y_surf, Z_surf, facecolors=m.to_rgba(slice_data),
                           shade=False, rstride=1, cstride=1, antialiased=False)
            ax.plot([0, box_len, box_len, 0, 0], [y_pos]*5, [0, 0, box_len, box_len, 0], 'k', lw=1)

        elif slice_plane == 'x':
            x_pos = slice_idx * (box_len / dim)
            slice_data = box_data[slice_idx, :, :].T
            Y_surf, Z_surf = np.meshgrid(coord, coord)
            X_surf = np.full_like(Y_surf, x_pos)

            ax.plot_surface(X_surf, Y_surf, Z_surf, facecolors=m.to_rgba(slice_data),
                           shade=False, rstride=1, cstride=1, antialiased=False)
            ax.plot([x_pos]*5, [0, box_len, box_len, 0, 0], [0, 0, box_len, box_len, 0], 'k', lw=1)
        # (Logic for X/Y planes omitted for brevity, but follows same pattern)

        # --- D. WIREFRAME CAGE ---
        corners = [0, box_len]
        for c1 in corners:
            for c2 in corners:
                ax.plot([0, box_len], [c1, c1], [c2, c2], 'gray', lw=0.5, alpha=0.3)
                ax.plot([c1, c1], [0, box_len], [c2, c2], 'gray', lw=0.5, alpha=0.3)
                ax.plot([c1, c1], [c2, c2], [0, box_len], 'gray', lw=0.5, alpha=0.3)

        # --- E. LABELS & VIEW ---
        ax.set_xlim(0, box_len); ax.set_ylim(0, box_len); ax.set_zlim(0, box_len)
        ax.set_xlabel('x [Mpc]'); ax.set_ylabel('y [Mpc]'); ax.set_zlabel('z [Mpc]')
        ax.set_title(f"z = {z_val:.2f} | Slice {slice_idx}")

        # Angle to see Floor, Back-Left (x=0), and Back-Right (y=Max) walls
        ax.view_init(elev=25, azim=-45)

        plt.colorbar(m, ax=ax, shrink=0.6, aspect=15, label=r'$\delta T_b$ [mK]')
        plt.show()

    # 3. CREATE SLIDERS
    # Z-Slider (Discrete values from your dictionary)
    z_slider = widgets.SelectionSlider(
        options=z_keys,
        value=z_keys[0],
        description='Redshift (z):',
        continuous_update=False # Prevents lag while dragging
    )

    # Slice Slider (Integer Index)
    # We assume all boxes have same dimensions, grab dim from first one
    first_box = list(data_dict.values())[0]
    max_slice = first_box.shape[0] - 1

    slice_slider = widgets.IntSlider(
        min=0, max=max_slice, step=1, value=max_slice//2,
        description='Slice Index:',
        continuous_update=False
    )

    # 4. LAUNCH
    interact(plot_frame, z_val=z_slider, slice_idx=slice_slider)

# --- HOW TO RUN ---
# Pass the entire dTb_array dictionary
visualize_4d_tomography(dTb_array, BOX_PARAM, slice_plane='x')
```

```python
#@title Cylindrically Averaged 2D Power Spectrum Slider
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import ipywidgets as widgets
from ipywidgets import interact
import numpy as np

# --- 1. PRE-CALCULATE DATA ---
# We store the results in a dictionary so the slider is instant.
print("Pre-calculating power spectra for all redshifts. Please wait...")
ps_cache = {}

for z in Z_PARAM_RANGE:
    # Run the PS calculation once per redshift
    ps, ksperp, kspar = t2c.power_spectrum_2d(
        dTb_array[z],
        # kbins=[k_bin_edges, k_bin_edges], # Uncomment if using custom bins
        box_dims=BOX_PARAM
    )
    ps_cache[z] = {
        'ps': ps,
        'ksperp': ksperp,
        'kspar': kspar
    }
print("Calculation complete! Launching slider...")

# --- 2. DEFINE UPDATE FUNCTION ---
def update_ps_plot(z_val):
    # Retrieve pre-calculated data
    data = ps_cache[z_val]
    ps = data['ps']
    ksperp = data['ksperp']
    kspar = data['kspar']

    # Dynamic vmin/vmax per redshift (as in your original code)
    vmin = np.nanmin(ps)
    vmax = np.nanmax(ps)

    # Safe checks for LogNorm
    if np.isnan(vmin) or vmin <= 0: vmin = 1e-5
    if np.isnan(vmax) or vmax <= 0: vmax = 1e5

    # Setup Plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # --- Left: 2D Heatmap ---
    mesh = ax1.pcolormesh(
        ksperp,
        kspar,
        ps.T,
        norm=LogNorm(vmin=vmin, vmax=vmax),
        cmap='Rainbow',
        shading='auto'
    )

    cbar = plt.colorbar(mesh, ax=ax1)
    cbar.set_label(r'$P(k)\ [\rm mK^2\ Mpc^3]$')

    ax1.set_xlabel(r'$k_\perp\ [\rm Mpc^{-1}]$')
    ax1.set_ylabel(r'$k_\parallel\ [\rm Mpc^{-1}]$')
    ax1.set_xscale('log')
    ax1.set_yscale('log')
    ax1.set_title(f'Cylindrical 2D PS (z={z_val})')

    # --- Right: 1D Slices ---
    # Safe indices that won't crash if the array is small
    slice_indices = [2, 7, 12]

    for idx in slice_indices:
        if idx < len(ksperp):
            k_perp_val = ksperp[idx]
            y_vals = ps[idx, :]
            mask = y_vals > 0

            if np.any(mask):
                ax2.loglog(kspar[mask], y_vals[mask], 'o-',
                           label=rf'$k_\perp \approx {k_perp_val:.2f}$')
            else:
                ax2.loglog([], [], 'o-',
                           label=rf'$k_\perp \approx {k_perp_val:.2f} (No Data)$')

    ax2.set_xlabel(r'$k_\parallel\ [\rm Mpc^{-1}]$')
    ax2.set_ylabel(r'$P(k_\perp, k_\parallel)$')
    ax2.set_title(r'1D Slices at fixed $k_\perp$')
    ax2.legend(loc='best', fontsize='small')
    ax2.grid(True, which="both", alpha=0.3)

    plt.tight_layout()
    plt.show()

# --- 3. CREATE SLIDER ---
# SelectionSlider snaps to your exact Z values
z_slider = widgets.SelectionSlider(
    options=sorted(list(ps_cache.keys()), reverse=True), # Sort high z to low z
    description='Redshift (z):',
    continuous_update=False # Prevents lagging while dragging
)

interact(update_ps_plot, z_val=z_slider)
```
